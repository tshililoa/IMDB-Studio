```python
# Use the code block to select movies that have been released between 2010-2016, with a rating of less than 6.0, but generated a significant (top) revenue.
#df1 = df.loc[(df['Year'] >= 2010, '<= 2016') & (df['Rating'] < 6.0) & (df['Revenue (Millions)'] > df['Revenue (Millions)'].quantile(0.95))] 
#df1 = data[((data['Year'] >= 2010) & (data['Year'] <= 2016))
 #           & (data['Rating'] < 6.0)
#          & (data['Revenue (Millions)'] > data['Revenue (Millions)'].quantile(0.95))]

df1 = df[((df['Year']   >= 2010) & (df['Year'] <= 2016)) 
            & (df['Rating'] < 6.0) 
            & (df['Revenue (Millions)'] > df['Revenue (Millions)'].quantile(0.95))] 
print(df1)
```
